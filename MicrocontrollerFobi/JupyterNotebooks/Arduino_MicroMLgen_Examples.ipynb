```python
# DECISIONTREE CLASSIFICATION, RAW
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from micromlgen import port

df = pd.read_csv('../Datasets/LegoTraining.csv',sep=',')
features = ['violet','blue','green','yellow','orange','red']
X = df[features]
y = df['label']
clf = DecisionTreeClassifier().fit(X, y)

predictions = clf.predict(X)
print("accuracy score: ", accuracy_score(y, predictions), "\n\n")

print(port(clf))
with open('decTreeClassifier.h', 'w') as file:
    file.write(port(clf))
```

```python
# RANDOMFOREST CLASSIFICATION, RAW
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from micromlgen import port

df = pd.read_csv('../Datasets/LegoTraining.csv',sep=',')
#print(df.head(5))
features = ['violet','blue','green','yellow','orange','red']
X = df[features]
y = df['label']
clf = RandomForestClassifier(n_estimators=3).fit(X, y)

predictions = clf.predict(X)
print("accuracy score: ", accuracy_score(y, predictions), "\n\n")

print(port(clf))
with open('RFClassifier.h', 'w') as file:
    file.write(port(clf))
```

```python
# SUPPORT VECTOR MACHINE, RAW
import pandas as pd
from sklearn.svm import SVC
from micromlgen import port
from sklearn.metrics import accuracy_score

df = pd.read_csv('../Datasets/LegoTraining.csv',sep=',')
features = ['violet','blue','green','yellow','orange','red']

X = df[features]
y = df['label']
clf = SVC(kernel='linear',gamma=0.001).fit(X,y)

predictions = clf.predict(X)
print("accuracy score: ", accuracy_score(y, predictions), "\n\n")

print(port(clf))
with open('linearSVM_Raw.h', 'w') as file:
    file.write(port(clf))
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from micromlgen import port

# Daten einlesen
df = pd.read_csv('../Datasets/LegoTraining.csv', sep=',')
features = ['violet', 'blue', 'green', 'yellow', 'orange', 'red']
X = df[features]
y = df['label']

# Modell erstellen und trainieren
clf = LogisticRegression(max_iter=1000).fit(X, y)

# Vorhersagen treffen
predictions = clf.predict(X)
print("accuracy score: ", accuracy_score(y, predictions), "\n\n")

# Modell exportieren
print(port(clf))
with open('logisticRegression.h', 'w') as file:
    file.write(port(clf))
```
